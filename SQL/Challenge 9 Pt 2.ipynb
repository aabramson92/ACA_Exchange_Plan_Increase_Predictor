```python
# What was the toal spent on salaries by each team, each year?
import pandas as pd
sql_query = '''SELECT teamid, yearid, SUM(salary) 
               FROM Salaries 
               GROUP BY teamid, yearid'''
pd.read_sql_query(sql_query, cnx)
```

```python
# What is the first and last year played on each team?
sql_query = '''SELECT playerid, MIN(yearid) AS first_year, MAX(yearid) AS last_year 
               FROM fielding 
               GROUP BY playerid;'''
pd.read_sql_query(sql_query, cnx)
```

```python
# Who has played the most all star games?
sql_query = '''SELECT CONCAT(namegiven, ' ', namelast) AS player, COUNT(*) AS asg_played 
               FROM allstarfull 
               JOIN master 
               ON allstarfull.playerid = master.playerid 
               GROUP BY namegiven, namelast 
               ORDER BY asg_played DESC'''
pd.read_sql_query(sql_query, cnx)
```

```python
sql_query = '''SELECT schoolid, COUNT(*) AS players 
               FROM schoolsplayers 
               GROUP BY schoolid 
               ORDER BY players DESC 
               LIMIT 1;'''
pd.read_sql_query(sql_query, cnx)
```

```python
sql_query = '''SELECT playerid, CONCAT(namegiven,' ',namelast) AS player, finalgame - debut AS days 
               FROM master;'''
pd.read_sql_query(sql_query, cnx)
```

```python
sql_query = '''SELECT EXTRACT(month from debut) AS debutmonth, COUNT(*) 
               FROM master 
               GROUP BY debutmonth 
               ORDER BY debutmonth;'''
pd.read_sql_query(sql_query, cnx)
```

```python
sql_query = '''SELECT salaries.playerid, AVG(salary) AS avg_salary
               FROM salaries
               LEFT JOIN master
               ON salaries.playerid = master.playerid 
               GROUP BY salaries.playerid
               ORDER BY avg_salary;'''
pd.read_sql_query(sql_query, cnx)
```
