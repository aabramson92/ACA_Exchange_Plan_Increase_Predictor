# 1) Import all Necessary Modules

```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn import naive_bayes
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import statsmodels.api as sm

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# 2) Merge all datasets for the analysis & modify features

```python
# Load insurance rate data and create dummy variables

df = pd.read_csv('insurance_data_v2.csv')
df['RateChange'] = (df['a.IndividualRate'] - df['IndividualRate'])/df['IndividualRate']
df['over15'] = np.where(df['RateChange'] >= 0.15, 1, 0)
df['bronze'] = np.where(df['MetalLevel_2015'] == 'Bronze', 1, 0)
df['silver'] = np.where(df['MetalLevel_2015'] == 'Silver', 1, 0)
df['gold'] = np.where(df['MetalLevel_2015'] == 'Gold', 1, 0)
df['platinum'] = np.where(df['MetalLevel_2015'] == 'Platinum', 1, 0)
df['MultistatePlan_2015'] = np.where(df['MultistatePlan_2015'] == 'Y', 1, 0)
```

```python
# Load population change data and merge with first dataframe

pg = pd.read_csv('popchange2015.csv')
pg['popchange'] = (pg.respop72015-pg.respop72014)/pg.respop72014 * 100
pg = pg[['FIPSCode', 'popchange']]
df1 = pd.merge(df, pg, on='FIPSCode')
```

```python
# Load population change data and merge with second dataframe

ai = pd.read_csv('county_age_data_15.csv')
ai['age50plus'] = ai['age5054'] + ai['age5559'] + ai['age60plus']
ai = ai[['FIPSCode', 'age50plus', 'median_age', 'adp']]
df2 = pd.merge(df1, ai, on='FIPSCode')
```

```python
# Load part 1 of county health metrics and merge with third dataframe

ch = pd.read_csv('health_data_pt_1.csv')
ch = ch[['FIPSCode', '% Fair/Poor', 'Physically Unhealthy Days', 'Mentally Unhealthy Days', '% Obese', '% Uninsured', '% Unemployed', 'Violent Crime Rate', 'Average Daily PM2.5', '% Severe Housing Problems', '% Long Commute - Drives Alone', 'PCP Rate', 'Dentist Rate', 'MHP Rate']]
ch = ch.fillna(value=0)
df3 = pd.merge(df2, ch, on='FIPSCode')
```

```python
# Load prior year's health insurance data, merge with fourth dataframe, and modify features

ts = pd.read_csv('insurance_time_series.csv')
ts['PlanId'] = ts['PlanId_2015']
ts = ts [['IndividualRate', 'PlanId', 'FIPSCode']]
df4 = pd.merge(df3, ts, on=['PlanId', 'FIPSCode'])
df4['PriorRateChange'] = (df4['IndividualRate_x'] - df4['IndividualRate_y'])/df4['IndividualRate_y']
df4['priorover20'] = np.where(df4['PriorRateChange'] >= 0.2, 1, 0)
```

```python
# Load part 2 of county health metrics and merge with fifth dataframe

hd = pd.read_csv('health_data_pt_2.csv')
hd = hd.fillna(value=0)
hd = hd[['FIPSCode', 'Age-Adjusted Mortality', '% Frequent Physical Distress', '% Frequent Mental Distress', '% Diabetic', 'HIV Prevalence Rate', '% Food Insecure',
       '% Limited Access', 'Drug Overdose Mortality Rate', '% Insufficient Sleep', 'Other PCP Rate', 'Household Income', 'Segregation Index',
       'Homicide Rate', '% Rural']]
df5 = pd.merge(df4, hd, on='FIPSCode')
```

```python
# Load health insurance plan features data, create dummy variables, and merge with sixth dataframe

pa = pd.read_csv('2015 Plan Data.csv')
pa = pa.fillna(value=0)
pa['PPO'] = np.where(pa['PlanType'] == 'PPO', 1, 0)
pa['HMO'] = np.where(pa['PlanType'] == 'HMO', 1, 0)
pa['POS'] = np.where(pa['PlanType'] == 'POS', 1, 0)
pa['Indemnity'] = np.where(pa['PlanType'] == 'Indemnity', 1, 0)
pa['EPO'] = np.where(pa['PlanType'] == 'EPO', 1, 0)
pa['IsHSAEligible'] = np.where(pa['IsHSAEligible'] == 'Yes', 1, 0)
pa['IsNewPlan'] = np.where(pa['IsNewPlan'] == 'New', 1, 0)
pa['IsReferralRequiredForSpecialist'] = np.where(pa['IsReferralRequiredForSpecialist'] == 'Yes', 1, 0)
pa['NationalNetwork'] = np.where(pa['NationalNetwork'] == 'Yes', 1, 0)
pa['OutOfServiceAreaCoverage'] = np.where(pa['OutOfServiceAreaCoverage'] == 'Yes', 1, 0)
pa['PlanId'] = pa['StandardComponentId']
df6 = pd.merge(df5, pa, on='PlanId')
```

# 3) Split values into y and X and conduct a Test-Train Split

```python
X = df6[['popchange', 'bronze', 'silver', 'gold', 'platinum', 'age50plus', 'median_age', 'adp', '% Fair/Poor', 'Physically Unhealthy Days', 'Mentally Unhealthy Days', '% Obese', '% Uninsured', '% Unemployed', 'Violent Crime Rate', 'Average Daily PM2.5', '% Severe Housing Problems', '% Long Commute - Drives Alone', 'PCP Rate', 'Dentist Rate', 'MHP Rate', 'priorover20', 'MultistatePlan_2015', 'Age-Adjusted Mortality', '% Frequent Physical Distress', '% Frequent Mental Distress', '% Diabetic', 'HIV Prevalence Rate', '% Food Insecure',
       '% Limited Access', 'Drug Overdose Mortality Rate', '% Insufficient Sleep', 'Other PCP Rate', 'Household Income', 'Segregation Index',
       'Homicide Rate', '% Rural', 'PPO', 'EPO', 'Indemnity', 'POS', 'HMO', 'IsHSAEligible', 'IsNewPlan', 'OutOfServiceAreaCoverage', 'IsReferralRequiredForSpecialist', 'NationalNetwork', 'BeginPrimaryCareCostSharingAfterNumberOfVisits']]
y = df6['over15']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# 4) Run Classification Models

```python
# Random Forest

randomforest = RandomForestClassifier(n_jobs=-1)
randomforest.fit(X_train, y_train)
y_predrf = randomforest.predict(X_test)
```

```python
# Gradient Boosting

gradientbooster = GradientBoostingClassifier()
gradientbooster.fit(X_train, y_train)
y_predgb = gradientbooster.predict(X_test)
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_predlog = logreg.predict(X_test)
```

```python
# K Nearest Neighbors

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_predknn = knn.predict(X_test)
```

```python
# Standard Vector Machine

svc = SVC()
svc.fit(X_train, y_train)
y_predsvc = svc.predict(X_test)
```

```python
# Naive Bayes

nb = naive_bayes.GaussianNB()
nb.fit(X_train, y_train)
y_prednb = nb.predict(X_test)
```

# 5) Compare Results of Classifications

```python
print("RANDOM FOREST")
print('AUC:       ', metrics.roc_auc_score(y_test, y_predrf))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_predrf))
print('Precision: ', metrics.precision_score(y_test, y_predrf))
print('F1 Score: ', metrics.f1_score(y_test, y_predrf))
print('Recall:    ', metrics.recall_score(y_test, y_predrf))
```

```python
print("GRADIENT BOOSTING")
print('AUC:       ', metrics.roc_auc_score(y_test, y_predgb))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_predgb))
print('Precision: ', metrics.precision_score(y_test, y_predgb))
print('F1 Score: ', metrics.f1_score(y_test, y_predgb))
print('Recall:    ', metrics.recall_score(y_test, y_predgb))
```

```python
print("LOGISTIC REGRESSION")
print('AUC:       ', metrics.roc_auc_score(y_test, y_predlog))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_predlog))
print('Precision: ', metrics.precision_score(y_test, y_predlog))
print('F1 Score: ', metrics.f1_score(y_test, y_predlog))
print('Recall:    ', metrics.recall_score(y_test, y_predlog))
```

```python
print("KNN")
print('AUC:       ', metrics.roc_auc_score(y_test, y_predknn))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_predknn))
print('Precision: ', metrics.precision_score(y_test, y_predknn))
print('F1 Score: ', metrics.f1_score(y_test, y_predknn))
print('Recall:    ', metrics.recall_score(y_test, y_predknn))
```

```python
print("SVC")
print('AUC:       ', metrics.roc_auc_score(y_test, y_predsvc))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_predsvc))
print('Precision: ', metrics.precision_score(y_test, y_predsvc))
print('F1 Score: ', metrics.f1_score(y_test, y_predsvc))
print('Recall:    ', metrics.recall_score(y_test, y_predsvc))
```

```python
print("NAIVE BAYES")
print('AUC:       ', metrics.roc_auc_score(y_test, y_prednb))
print('Accuracy:  ', metrics.accuracy_score(y_test, y_prednb))
print('Precision: ', metrics.precision_score(y_test, y_prednb))
print('F1 Score: ', metrics.f1_score(y_test, y_prednb))
print('Recall:    ', metrics.recall_score(y_test, y_prednb))
```

Since Random Forest has the best AUC curve and the best Precision and Recall, that is the model that will be used.

NB: Were none of these models as precise as they are, further feature modification would have been conducted, such as normalizing the features or over-sampling the minority class using SMOTE or another similar tactic.

# 6) Graphics of Results

```python
# Create AUC Scores

y_scorerf = randomforest.predict_proba(X_test)[::,1]
y_scorenb = nb.predict_proba(X_test)[::,1]
y_scorelog = logreg.predict_proba(X_test)[::,1]
fprrf, tprrf,_ = metrics.roc_curve(y_test, y_scorerf)
fprnb, tprnb,_ = metrics.roc_curve(y_test, y_scorenb)
fprlog, tprlog,_ = metrics.roc_curve(y_test, y_scorelog)
aucrf = metrics.roc_auc_score(y_test, y_scorerf)
aucnb = metrics.roc_auc_score(y_test, y_scorenb)
auclog = metrics.roc_auc_score(y_test, y_scorelog)
plt.plot(fprnb,tprnb,label="Naive Bayes")
plt.plot(fprrf,tprrf,label="Random Forest")
plt.plot(fprlog,tprlog,label="Logistic Regression")
plt.legend(loc=4)
plt.title("ROC curves")
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.show();
```

```python
# Plot Confusion Matrix

def plot_confusion_matrix(cm,title='Confusion matrix', cmap=plt.cm.Reds):
    plt.imshow(cm, interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.xticks(range(cm.shape[1]))
    plt.yticks(range(cm.shape[0]))
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.savefig('confusion_matrix.svg')
cm = metrics.confusion_matrix(y_test, y_predrf, labels=None)
plot_confusion_matrix(cm)
```
